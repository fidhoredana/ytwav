```python
#@title Download Youtube to WAV
from __future__ import unicode_literals
import yt_dlp
import ffmpeg
import sys
import urllib.request, json

# retrieve video title using urllib.request
url = "https://youtu.be/JYstE5GwT2U" #@param {type:"string"}
with urllib.request.urlopen(f"https://www.youtube.com/oembed?url={url}&format=json") as response:
    data = json.load(response)
    judul = data['title']

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }],
    "outtmpl": f'drive/MyDrive/Youtube-Music/{judul}',
}

def download_from_url(url, judul):
    ydl.download([url])

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    download_from_url(url, judul)
```
