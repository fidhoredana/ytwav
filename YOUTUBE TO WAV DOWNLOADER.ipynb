```python
#@title Download Youtube to WAV
from __future__ import unicode_literals
import yt_dlp
import ffmpeg
import sys
import urllib.request, json

# retrieve video title using urllib.request
url = "" #@param {type:"string"}
with urllib.request.urlopen(f"https://www.youtube.com/oembed?url={url}&format=json") as response:
    data = json.load(response)
    judul = data['title']

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }],
    "outtmpl": f'drive/MyDrive/Youtube-Music/{judul}',
}

def download_from_url(url, judul):
    ydl.download([url])

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    download_from_url(url, judul)
```

```python
#@title Download Youtube to WAV Bulk
# @markdown Note:
# @markdown  Run the cell first, then a tab to insert URL will shown (1 line for 1 URL, separated with new line), click "Download WAV" button to start!
from __future__ import unicode_literals
import yt_dlp
import ffmpeg
import urllib.request, json
import ipywidgets as widgets
from IPython.display import display

# Global variable for ydl_opts
ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }],
    "outtmpl": 'drive/MyDrive/Youtube-Music/%(title)s.%(ext)s',
}

# Function to download videos from URLs
def download_videos_from_urls(urls):
    for url in urls:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            download_from_url(url.strip())

# Function to download a single video from a URL
def download_from_url(url):
    with urllib.request.urlopen(f"https://www.youtube.com/oembed?url={url}&format=json") as response:
        data = json.load(response)
        judul = data['title']

    ydl_opts['outtmpl'] = f'drive/MyDrive/Youtube-Music/{judul}.%(ext)s'

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

# Create textarea widget for user input
textarea = widgets.Textarea(
    placeholder='Enter YouTube URLs here (one per line)',
    description='YouTube URLs:',
    layout=widgets.Layout(width='500px', height='200px')
)

# Create a button widget
button = widgets.Button(description="Download WAV")

# Function to handle button click event
def handle_button_click(b):
    urls_input = textarea.value.strip()
    urls = urls_input.split('\n')
    download_videos_from_urls(urls)
    print("Videos downloaded successfully!")

# Attach the handle_button_click function to the button's click event
button.on_click(handle_button_click)

# Display the widgets
display(textarea)
display(button)
```
